subplot = Matplotlib allows managment of multiple figures simultaneously , and with each figure it offeres the ability to view different plots defined as subplots.The subplot() function in addition to sub dividing the figure in different drawing areas is used at the same time in order to focus the commands on a specific subplot . subplot() function sets the mode of subdivision and which is the current subplot. The subplot() function is composed of three integers , The first number defines how many parts the figure is split  into vertically . The second number defines how many parts the figure is divided into horizontally . The third issue selects which is the current subplot on which i can direct commands.

```python
t = np.arange(0,5,0.1)
y1 = np.sin(2*np.pi*t)
y2 = np.sin(2*np.pi*t)
```

```python
plt.subplot(211)#devide the canvas into two horizontal subplots
plt.plot(t,y1,'b-.')
plt.subplot(212)
plt.plot(t,y2,'r--')
```

```python
plt.show()
```

Manageing Dates is often required while createing a graph , To handle the Date manipulation use time scale with appropiate objects . First i need to import matplotlib.dates , this module is specialized in date manipulation in graph axis .Use MonthLocator() and DayLocator() function to scale the days and months.  

```python
import datetime
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
months = mdates.MonthLocator()
days = mdates.DayLocator()
timFmt = mdates.DateFormatter('%Y-%m')
events = [datetime.date(2015,1,23),datetime.date(2015,1,28),datetime.date(2015,2,3)
         ,datetime.date(2015,2,21),datetime.date(2015,3,15),datetime.date(2015,3,24),
         datetime.date(2015,4,8),datetime.date(2015,4,20),datetime.date(2015,5,10),
         datetime.date(2015,5,29)]
readings = [12,23,34,4,10,20,20,36,37,40]
fig,ax = plt.subplots()
plt.plot(events,readings)
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timFmt)
ax.xaxis.set_minor_locator(days)
```

```python
plt.show()
```

use of xticks() and yticks() function which will helps to define the axis according to the name

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-2*np.pi,2*np.pi,0.01)
y = np.sin(3*x)/x
y2= np.sin(2*x)/x
y3= np.sin(x)/x
plt.plot(x,y,color='b')
plt.plot(x,y2,color='r')
plt.plot(x,y3,color='g')
plt.xticks([-2*np.pi,-np.pi,0,np.pi,2*np.pi],[r'$-2\pi$',r'$-\pi$',r'$0$',r'$+\pi$',r'$+2\pi$'])
plt.yticks([-1,0,+1,+2,+3],[r'$-1$',r'$0$',r'$+1$',r'$+2$',r'$+3$'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-2*np.pi,2*np.pi,0.01)
y = np.sin(3*x)/x
y2= np.sin(2*x)/x
y3= np.sin(x)/x
plt.plot(x,y,color='b')
plt.plot(x,y2,color='r')
plt.plot(x,y3,color='g')
plt.xticks([-2*np.pi,-np.pi,0,np.pi,2*np.pi],[r'$-2\pi$',r'$-\pi$',r'$0$',r'$+\pi$',r'$+2\pi$'])
plt.yticks([-1,0,+1,+2,+3],[r'$-1$',r'$0$',r'$+1$',r'$+2$',r'$+3$'])
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```

Another specific arguments can be passed in the bar() function , For Example i can add standard deviation values of the bar through the yerr kwarg along with a list containing standard deviation , error_kw which accepts other kwargs for representing error bars, eColor which specifies the color of the bars , and capsize which defines the width of the traverse lines that marks the end of the error bars . Another kwarg is alpha , which indicates the degree of transperency of the colored bar . Alpha is a value ranging from 0 to 1 . 

```python
import numpy as np
index = np.arange(5)
values = [5,7,3,4,6]
std = [0.8,1,0.4,0.9,1.3]
plt.title('Bar Chart with Standard Deviation')
plt.bar(index,values,yerr=std,error_kw={'ecolor':'0.1',
                                       'capsize':6},alpha=0.7,label='First')
plt.xticks(index+0.4,['A','B','C','D','E'])
```

```python
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
index = np.arange(5)
values1 = [5,7,3,4,6]
values2 = [6,6,4,5,7]
values3 = [5,6,5,4,6]
bw = 0.3
plt.axis([0,5,0,8])
plt.title('MultiSeries Bar Chart')
plt.bar(index,values1,bw,color='b')
plt.bar(index+bw,values2,color='g')
plt.bar(index+2*bw,values3,color='r')
plt.xticks(index+1.5*bw,['A','B','C','D','E'])
```
